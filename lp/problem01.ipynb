$$

g_3(\mathbf{x})&=x_1-x_2\ge-8=b_3\\
g_4(\mathbf{x})&=x_1-x_2\le8=b_4\\

$
\begin{aligned}

v_2&=\begin{bmatrix}1\\-1\\\end{bmatrix}\,\,\,\text{points SE}\,\,\,&;\,\,\,&\nabla z\cdot v_2=3\cdot1+2\cdot-1&=3-2=1\gt0\,\,\,\text{points SE}\\
v_3&=\begin{bmatrix}1\\ 1\\\end{bmatrix}\,\,\,\text{points NE}\,\,\,&;\,\,\,&\nabla z\cdot v_3=3\cdot1+2\cdot 1&=3+2=5\gt0\,\,\,\text{points NE}\\

\end{aligned}
$
